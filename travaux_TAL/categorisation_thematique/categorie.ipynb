```python
from bs4 import BeautifulSoup
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import pyLDAvis
import pyLDAvis.sklearn
pyLDAvis.enable_notebook()
```

```python
with open('ap/ap.txt',"r") as f:
    soup = BeautifulSoup(f.read(),"xml")
    #stemmer = SnowballStemmer("french", ignore_stopwords=True)        
    
corpus = np.array([ d.text   for d in soup.find_all("TEXT")])
```

```python
vectorizer = CountVectorizer(stop_words='english',token_pattern=r'[a-zA-Z]+')
Xvec = vectorizer.fit_transform(corpus)

feature_n = vectorizer.get_feature_names()
```

```python
n = 40 # nb de clusters

lda = LatentDirichletAllocation(n, max_iter=5,
                                learning_method='online',
                                learning_offset=50.,
                                random_state=1)

lda.fit(Xvec)
```

```python
pyLDAvis.sklearn.prepare(lda,Xvec,vectorizer)
```
